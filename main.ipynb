```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def rect(x):
    if abs(x) > 1/2:
        return 0
    elif abs(x) == 1/2:
        return 1/2
    else:
        return 1
def tri(x):
    if abs(x) < 1:
        return 1 - abs(x)
    else:
        return 0
```

```python
def y(x):
    if  x < 0 or x > 8:
        return 0
    elif 0 <= x <1:
        return 1
    elif 1 <= x < 3:
        return x
    elif 3 <= x < 4:
        return 3
    elif 4 <= x < 5:
        return x-1
    else:
        return -x+9
```

```python
x = np.arange(-10., 10., 0.01)
plt.plot(x, [rect(i) for i in x])
```

```python
plt.plot(x, [tri(i) for i in x])
```

```python
plt.plot(x, [y(i) for i in x])
```

```python
def f(x):
    return rect(x-1/2) + rect((x - 4.5)/7)+tri(x-5)+tri((x-3)/2)*2+tri((x-5)/2)*2+tri(x-6)#+tri(x-7)
plt.plot(x, [f(i) for i in x])
```

```python
def g(x):
    return y(x)-f(x)
print(g(7))
plt.plot(x, [g(i) for i in x])
plt.axis((-1, 10, -1, 4))
```
